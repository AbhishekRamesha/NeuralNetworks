```
def read_files(filename):
  with open(filename, 'rb') as f:
    zero, data_type, dims = struct.unpack('>HBB' , f.read(4))
    shape = tuple(struct.unpack('>I' , f.read(4))[0] for d in range(dims))
    return np.frombuffer(f.read(), dtype=np.uint8).reshape(shape)
```

```
#importing files 
train_images = read_files('/content/drive/My Drive/Colab Notebooks/HW2/train-images.idx3-ubyte')
train_label = read_files('/content/drive/My Drive/Colab Notebooks/HW2/train-labels.idx1-ubyte')
test_images = read_files('/content/drive/My Drive/Colab Notebooks/HW2/t10k-images.idx3-ubyte')
test_label = read_files('/content/drive/My Drive/Colab Notebooks/HW2/t10k-labels.idx1-ubyte')
```

```
def activation(x):
  i = 0
  y = np.empty([10,1])
  for j in x:
    if j >= 0:
      y[i] = 1.0
    else:
      y[i] = 0.0
    i += 1
  return y
```

```
def train_err(epoch, error):
  for i in range(n):
    xi = train_images[i]
    xi.resize(784, 1)
    a = np.matmul(w,xi)
    predict = a.argmax(axis=0)
    actual = train_label[i]
    if predict != actual:
      error[epoch] += 1
  return error[epoch]
```

```
def update_weights(w):
  for i in range(n):
    xi = train_images[i]
    xi.resize(784, 1)
    y = np.array(activation(np.matmul(w,xi)))
    lab = np.zeros((1,10)).T
    lab[train_label[i]] = 1
    diff = lab - y
    xib = np.transpose(xi)
    update = learn_rate * np.matmul(diff, xib)
    w += update
```

```
def learn_weight(w, epoch, thold, learn_rate):
  while epoch<100:
    error.append(0)
    error[epoch] = train_err(epoch, error)
    update_weights(w)
    epoch += 1
    if error[epoch-1]/n <= thold:
      break
```

```
def find_err():
  test_err = 0
  for i in range(len(test_images)):
    xi = test_images[i]
    xi.resize(784, 1)
    a = np.matmul(w, xi)
    predict = a.argmax(axis=0)
    actual = test_label[i]
    if predict != actual:
      test_err += 1
  print("Number of errors in test images: ", test_err)
  print("Percentage of errors in test images: ",test_err*100/len(test_images) )
  
```

```
def plot(error):
  fig, ax = plt.subplots(figsize=(10,10))
  plt.plot(range(len(error)), error, c='red')
  plt.ylabel('Number of Misclassifications')
  plt.xlabel('Number of Epochs')
  plt.title('Error vs Epoch')
  plt.show()
```

```
w = np.random.uniform(-1, 1, size=(10,784))
n = 50
epoch = 0
thold = 0.0
learn_rate = 1.0
error = []
learn_weight(w, epoch, thold, learn_rate)
plot(error)
find_err()
```
